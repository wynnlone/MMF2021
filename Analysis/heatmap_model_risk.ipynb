```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile('heatmap_use.xls')
```

```python
df1 = pd.read_excel(xls, '2006_2011')
df2 = df1.iloc[:,1:]

df3 = pd.read_excel(xls, '2011_2016')
df4 = df3.iloc[:,1:]

df5 = pd.read_excel(xls, '2016_2021')
df6 = df5.iloc[:,1:]

df7 = pd.read_excel(xls, '2021 YTD')
df8 = df7.iloc[:,1:]
```

```python
corr_heatmap = sns.heatmap(df2.corr(),vmin = -1, vmax = 1,cmap='YlGnBu');
corr_heatmap.set_xticklabels(
    corr_heatmap.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.savefig('corr_heatmap_2006_2011.png', dpi=300, bbox_inches='tight')
```

```python
corr_heatmap = sns.heatmap(df4.corr(),vmin = -1, vmax = 1,cmap='YlGnBu');
corr_heatmap.set_xticklabels(
    corr_heatmap.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.savefig('corr_heatmap_2011_2016.png', dpi=300, bbox_inches='tight')
```

```python
corr_heatmap = sns.heatmap(df6.corr(),vmin = -1, vmax = 1,cmap='YlGnBu');
corr_heatmap.set_xticklabels(
    corr_heatmap.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.savefig('corr_heatmap_2016_2021.png', dpi=300, bbox_inches='tight')
```

```python
corr_heatmap = sns.heatmap(df8.corr(),vmin = -1, vmax = 1,cmap='YlGnBu');
corr_heatmap.set_xticklabels(
    corr_heatmap.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.savefig('corr_heatmap_2021 YTD.png', dpi=300, bbox_inches='tight')
```
